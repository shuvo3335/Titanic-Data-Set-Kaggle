```python
## Categorical
def bar_chartFig(feature):
    survived_people = train[train['Survived']==1] [feature].value_counts()
    dead_people = train[train['Survived']==0] [feature].value_counts()
    df = pd.DataFrame([survived_people,dead_people])
    df.index = ['Survived','Dead']
    df.plot(kind = 'bar', stacked = True, figsize = (10,5))
```

```python
bar_chartFig("Sex")
```

```python
sns.countplot(x = 'Survived',hue = 'Sex', data = train)
plt.title('male female survived level')
plt.show()
```

```python
bar_chartFig('Pclass')
```

```python
bar_chartFig('Parch')
```
